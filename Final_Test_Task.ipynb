```python
# Question 8
df2 = pd.read_csv('Данные для тестового задания - Данные АБ тестов.csv')
df2.head()
```

```python
df2.info()
```

```python
from scipy.stats import ttest_ind
results = []

for num in df2["experiment_num"].unique():
    data = df2[df2["experiment_num"] == num]
    test = data[data["experiment_group"] == "test"]["revenue"]
    control = data[data["experiment_group"] == "control"]["revenue"]
    t_stat, p_value = ttest_ind(test, control, equal_var = False)
    results.append((num, test.mean(), control.mean(), p_value))
results_df = pd.DataFrame(results, columns = ["experiment", "test_mean", "control_mean", "p_val"])
results_df
```

```python
# Question 18
cr_control = 1003 / 100047501
cr_test = 1099 / 100001055
(cr_test - cr_control) * 100
```

```python
from statsmodels.stats.proportion import proportions_ztest
results = [1003, 1099]
users = [100047501, 100001055]

z_stat, p_value = proportions_ztest(results, users)
print(f"z_stat: {z_stat}")
print(f"p-value: {p_value}")
```

```python
from statsmodels.stats.proportion import confint_proportions_2indep

ci_low, ci_high = confint_proportions_2indep(count1 = 1003, nobs1 = 100047501, count2 = 1099, nobs2 = 100001055, method = "wald")
print(f"Confidence interval: {ci_low} - {ci_high}")
```
